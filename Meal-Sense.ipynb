## 1. Data Generation

```python
import pandas as pd
import numpy as np

# Generate synthetic dataset
np.random.seed(42)

data = {
    "student_id": np.arange(1, 201),
    "school_id": np.random.randint(1, 10, 200),
    "attendance": np.random.randint(50, 100, 200),
    "meals_served": np.random.randint(40, 100, 200),
    "food_quality_score": np.random.randint(1, 10, 200),
    "complaints_count": np.random.randint(0, 10, 200)
}

df = pd.DataFrame(data)

df.head()
```

### Observations

- The dataset contains student-level and school-level information.

```python
import pandas as pd

# ── Load cleaned dataset ──────────────────────────────────────────────────────

print(f"Rows loaded: {len(df)}\n")

# ═══════════════════════════════════════════════════════════════════════════════
# ANALYSIS 1 — Average meals served per school
# Groups all rows by school and calculates the mean meals served.
# Helps identify which schools serve the most and least food.
# ═══════════════════════════════════════════════════════════════════════════════

avg_meals = (
    df.groupby("school_id")["meals_served"]
    .mean()
    .round(2)
    .sort_values(ascending=False)
    .reset_index()
)
avg_meals.columns = ["school_id", "avg_meals_served"]

print("=" * 45)
print("AVERAGE MEALS SERVED PER SCHOOL")
print("=" * 45)
print(avg_meals.to_string(index=False))

# ═══════════════════════════════════════════════════════════════════════════════
# ANALYSIS 2 — Average attendance per school
# Shows how many students turn up on average at each school.
# Compare with meals served to spot over/under-serving.
# ═══════════════════════════════════════════════════════════════════════════════

avg_attendance = (
    df.groupby("school_id")["attendance"]
    .mean()
    .round(2)
    .sort_values(ascending=False)
    .reset_index()
)
avg_attendance.columns = ["school_id", "avg_attendance"]

print("\n" + "=" * 45)
print("AVERAGE ATTENDANCE PER SCHOOL")
print("=" * 45)
print(avg_attendance.to_string(index=False))

# ═══════════════════════════════════════════════════════════════════════════════
# ANALYSIS 3 — Complaints distribution
# Looks at how complaints are spread across all days.
# A right-skewed distribution means most days are fine but a few are very bad.
# ═══════════════════════════════════════════════════════════════════════════════

print("\n" + "=" * 45)
print("COMPLAINTS DISTRIBUTION")
print("=" * 45)
print(df["complaints_count"].describe().round(2))

print("\nComplaints breakdown (count of days per complaint level):")
print(df["complaints_count"].value_counts().sort_index().to_string())

# ═══════════════════════════════════════════════════════════════════════════════
# ANALYSIS 4 — Correlation matrix
# Measures how strongly each numeric column is related to the others.
# Values range from -1 (opposite) to +1 (identical movement).
# 0 means no relationship at all.
# ═══════════════════════════════════════════════════════════════════════════════

cols = ["attendance", "meals_served", "food_quality_score", "complaints_count"]
corr = df[cols].corr().round(3)

print("\n" + "=" * 45)
print("CORRELATION MATRIX")
print("=" * 45)
print(corr.to_string())

# ── Key correlation insights ──────────────────────────────────────────────────
print("\n" + "=" * 45)
print("KEY INSIGHTS FROM CORRELATION")
print("=" * 45)

pairs = []
for i in range(len(cols)):
    for j in range(i + 1, len(cols)):
        pairs.append((cols[i], cols[j], corr.iloc[i, j]))

pairs.sort(key=lambda x: abs(x[2]), reverse=True)

for col1, col2, val in pairs:
    if abs(val) >= 0.4:
        strength = "strong"
    elif abs(val) >= 0.2:
        strength = "moderate"
    else:
        strength = "weak / no"
    direction = "positive" if val > 0 else "negative"
    print(f"  {col1} vs {col2}: {val:+.3f}  →  {strength} {direction} relationship")
```

## 4. Anomaly Detection

```python
# Detect shortages (meals < attendance)
shortage = df[df["meals_served"] < df["attendance"]]

print("SHORTAGE CASES (meals < attendance)")
print(shortage.head())


# Detect over-supply (meals > attendance)
oversupply = df[df["meals_served"] > df["attendance"]]

print("\nOVER-SUPPLY CASES (meals > attendance)")
print(oversupply.head())


# Low food quality
low_quality = df[df["food_quality_score"] < 4]

print("\nLOW FOOD QUALITY CASES")
print(low_quality.head())


# High complaints
high_complaints = df[df["complaints_count"] > 6]

print("\nHIGH COMPLAINT CASES")
print(high_complaints.head())
```

### Anomaly Insights

- Several cases show meals served are less than attendance, indicating possible food shortages.
- Some cases show meals served are more than attendance, indicating potential wastage.
- Low food quality scores are observed in multiple entries, which may lead to dissatisfaction.
- High complaint counts indicate possible hygiene or quality issues.

These anomalies highlight inefficiencies in the mid-day meal system and areas that require attention.

```python
import pandas as pd

# ── Load cleaned dataset ──────────────────────────────────────────────────────
# STEP 1 — Estimate required meals based on attendance

df["recommended_meals"] = df["attendance"] * 1.05
print(f"Rows loaded: {len(df)}\n")

# ═══════════════════════════════════════════════════════════════════════════════
# STEP 1 — Estimate required meals based on attendance
#
# Rule: Every present student needs exactly 1 meal.
#       We also add a 5% buffer for safety (late arrivals, miscounts, spillage).
#
#   required_meals    = attendance          ← bare minimum, no student goes without
#   recommended_meals = attendance × 1.05  ← safe target with buffer
# ═══════════════════════════════════════════════════════════════════════════════

BUFFER = 0.05   # 5% safety margin

df["required_meals"]    = df["attendance"]
df["recommended_meals"] = (df["attendance"] * (1 + BUFFER)).astype(int)

print("=" * 55)
print("STEP 1 — MEAL ESTIMATES")
print("=" * 55)
print(f"Buffer used           : {int(BUFFER * 100)}%")
print(f"Avg attendance        : {df['attendance'].mean():.1f}")
print(f"Avg required meals    : {df['required_meals'].mean():.1f}")
print(f"Avg recommended meals : {df['recommended_meals'].mean():.1f}")
print(f"Avg meals served      : {df['meals_served'].mean():.1f}")

# ═══════════════════════════════════════════════════════════════════════════════
# STEP 2 — Detect over-supply and under-supply
#
# We compare meals_served against the two thresholds:
#
#   over-supply  → meals_served > recommended_meals  (wasteful)
#   under-supply → meals_served < required_meals     (students missed a meal)
#   optimal      → required <= meals_served <= recommended  (just right)
# ═══════════════════════════════════════════════════════════════════════════════

def classify_supply(row):
    if row["meals_served"] > row["recommended_meals"]:
        return "over-supply"
    elif row["meals_served"] < row["required_meals"]:
        return "under-supply"
    else:
        return "optimal"

df["supply_status"] = df.apply(classify_supply, axis=1)

# Gap = meals_served − recommended  (negative = shortage, positive = excess)
df["gap"] = df["meals_served"] - df["recommended_meals"]

status_counts = df["supply_status"].value_counts()

print("\n" + "=" * 55)
print("STEP 2 — SUPPLY STATUS ACROSS ALL DAYS")
print("=" * 55)
print(f"  Optimal days      : {status_counts.get('optimal', 0)}")
print(f"  Under-supply days : {status_counts.get('under-supply', 0)}")
print(f"  Over-supply days  : {status_counts.get('over-supply', 0)}")

total_shortage = df.loc[df["gap"] < 0, "gap"].abs().sum()
total_excess   = df.loc[df["gap"] > 0, "gap"].sum()

print(f"\n  Total meals short  (under-supply): {int(total_shortage)}")
print(f"  Total excess meals (over-supply) : {int(total_excess)}")
print(f"  Avg daily gap                    : {df['gap'].mean():.1f} meals")

# ═══════════════════════════════════════════════════════════════════════════════
# STEP 3 — Per-school anomaly report
#
# For each school, summarise:
#   • how often supply was short or excess
#   • average daily deficit (meals missing below the recommended target)
#   • suggested daily allocation going forward
# ═══════════════════════════════════════════════════════════════════════════════

df["daily_deficit"] = (df["recommended_meals"] - df["meals_served"]).clip(lower=0)
df["daily_excess"]  = (df["meals_served"] - df["recommended_meals"]).clip(lower=0)

school_report = (
    df.groupby("school_id")
    .agg(
        avg_attendance    = ("attendance",       "mean"),
        avg_meals_served  = ("meals_served",     "mean"),
        avg_recommended   = ("recommended_meals","mean"),
        avg_daily_deficit = ("daily_deficit",    "mean"),
        avg_daily_excess  = ("daily_excess",     "mean"),
        under_supply_days = ("supply_status",    lambda x: (x == "under-supply").sum()),
        over_supply_days  = ("supply_status",    lambda x: (x == "over-supply").sum()),
        optimal_days      = ("supply_status",    lambda x: (x == "optimal").sum()),
    )
    .round(2)
    .reset_index()
)

print("\n" + "=" * 55)
print("STEP 3 — PER-SCHOOL ANOMALY REPORT")
print("=" * 55)
print(school_report.to_string(index=False))

# ═══════════════════════════════════════════════════════════════════════════════
# STEP 4 — Suggest improved meal allocation per school
#
# Strategy:
#   Look at each school's historical average attendance.
#   Add a 5% buffer.  Round up so no student is ever short a meal.
#   Flag schools with the worst under-supply for priority action.
# ═══════════════════════════════════════════════════════════════════════════════

allocation = school_report[["school_id","avg_attendance","avg_meals_served","avg_recommended","avg_daily_deficit"]].copy()
allocation["suggested_daily_allocation"] = (allocation["avg_attendance"] * (1 + BUFFER)).round().astype(int)
allocation["meals_to_add_per_day"]       = (allocation["suggested_daily_allocation"] - allocation["avg_meals_served"]).round(1)

# Rank schools by urgency (largest deficit first)
allocation = allocation.sort_values("avg_daily_deficit", ascending=False).reset_index(drop=True)
allocation.index += 1   # start ranking from 1

print("\n" + "=" * 55)
print("STEP 4 — SUGGESTED DAILY ALLOCATION PER SCHOOL")
print("=" * 55)
print(allocation.to_string())

# ═══════════════════════════════════════════════════════════════════════════════
# STEP 5 — Flag worst-performing days for investigation
#
# Days where shortage is largest are the most urgent.
# Print the 10 worst days across all schools.
# ═══════════════════════════════════════════════════════════════════════════════

worst_days = (
    df[df["supply_status"] == "under-supply"]
    [["school_id","attendance","meals_served","recommended_meals","gap","food_quality_score","complaints_count"]]
    .sort_values("gap")        # most negative gap = worst shortage
    .head(10)
    .reset_index(drop=True)
)
worst_days.index += 1

print("\n" + "=" * 55)
print("STEP 5 — TOP 10 WORST UNDER-SUPPLY DAYS")
print("=" * 55)
print(worst_days.to_string())

# ═══════════════════════════════════════════════════════════════════════════════
# STEP 6 — System-wide efficiency summary
# ═══════════════════════════════════════════════════════════════════════════════

current_total   = df["meals_served"].sum()
optimised_total = df["suggested_daily_allocation"] if "suggested_daily_allocation" in df.columns else df["recommended_meals"].sum()
total_recommended = df["recommended_meals"].sum()
efficiency = (current_total / total_recommended * 100).round(1)

print("\n" + "=" * 55)
print("STEP 6 — EFFICIENCY SUMMARY")
print("=" * 55)
print(f"Total meals currently served  : {current_total:,}")
print(f"Total meals recommended       : {total_recommended:,}")
print(f"Supply efficiency             : {efficiency}%")
print(f"Meals gap to close            : {total_recommended - current_total:,}")
print(f"\nAction: Increase daily allocation by ~{(total_recommended - current_total) / df['school_id'].nunique() / df.shape[0] * df['school_id'].nunique():.0f} meals per school per day on average.")

# ── Save report ───────────────────────────────────────────────────────────────
df.to_csv("midday_meal_optimised.csv", index=False)
school_report.to_csv("school_supply_report.csv", index=False)

print("\n💾 Saved: midday_meal_optimised.csv")
print("💾 Saved: school_supply_report.csv")
```
